```python
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
import nltk
import string
nltk.download('stopwords')
from nltk.corpus import stopwords
stops=set(stopwords.words('english'))
punctuations=list(string.punctuation)
stops.update(punctuations)
```

```python
def clear_review(words):
    output_words=[]
    for w in words:
        if w.lower() not in stops:
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents=[(clear_review(document),category) for document, category in document]
x_test1=[(clear_review(x_test1)) for x_test1 in x_test1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
y_train = [category for document, category in documents]
x_train = [" ".join(document) for document, category in documents]
```

```python
xtest=[" ".join(x_test1) for x_test1 in x_test1]
count_vec = CountVectorizer(max_features = 2000, ngram_range=(1,2))
x_train_features = count_vec.fit_transform(x_train)
x_train_features.todense()
```
